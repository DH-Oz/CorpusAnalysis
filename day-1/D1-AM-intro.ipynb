```python
import matplotlib.pyplot as plt

token_counts = []
for text in df["text"]:
    token_counts.append(len(text.split()))

df_plot = df.copy()
df_plot["tokens"] = token_counts

fig, ax = plt.subplots(figsize=(13, 5))
for party in df_plot["party"].unique():
    sub = df_plot[df_plot["party"] == party].sort_values("year")
    ax.plot(sub["year"], sub["tokens"], marker="o", linewidth=0.6, markersize=4, label=party)
ax.set_xlabel("Year")
ax.set_ylabel("Tokens (whitespace-split)")
ax.set_title("State of the Union address length, 1790-2026")
ax.legend(loc="upper left", fontsize=9, ncol=2)
plt.tight_layout()
plt.show()
```

```python
import sklearn.feature_extraction.text
import pandas

labels = []
for year, pres in zip(df["year"], df["president"]):
    labels.append(f"{year}-{pres}")

# Build the document-feature matrix, then trim rare words: keep a term only if
# it appears in at least 3 different speeches (min_df=3) and at least 5 times in
# total.
vectorizer = sklearn.feature_extraction.text.CountVectorizer(lowercase=True, stop_words="english", min_df=3)
sparse_counts = vectorizer.fit_transform(df["text"])
features = vectorizer.get_feature_names_out()

counts = sparse_counts.toarray()
column_totals = counts.sum(axis=0)
keep = column_totals >= 5
counts = counts[:, keep]
features = features[keep]

dfm_df = pandas.DataFrame(counts, index=labels, columns=features)
print(f"DFM shape after trimming: {counts.shape[0]} docs x {counts.shape[1]} features")
print("Head (first 6 docs, first 8 features):")
print(dfm_df.iloc[:6, :8])

totals = dfm_df.sum(axis=0).sort_values(ascending=False)
print("Top 20 features across the whole corpus:")
print(totals.head(20))
```

<!-- demo slide -->

```python
import wordcloud

word_freq = {}
for word, total in totals.items():
    word_freq[word] = int(total)

wc = wordcloud.WordCloud(width=900, height=500, background_color="white", colormap="tab10", random_state=42)
wc.generate_from_frequencies(word_freq)

fig, ax = plt.subplots(figsize=(10, 5))
ax.imshow(wc, interpolation="bilinear")
ax.axis("off")
ax.set_title("State of the Union — top terms (1790–2026)")
plt.tight_layout()
plt.show()
```

## Topic modeling using Latent Dirichlet Allocation

```python
import sklearn.decomposition

n_topics = 8
n_terms = 10
lda = sklearn.decomposition.LatentDirichletAllocation(n_components=n_topics, random_state=42, max_iter=20)
lda.fit(counts)

topics = {}
for topic_idx, topic_weights in enumerate(lda.components_):
    top_term_indices = topic_weights.argsort()[-n_terms:][::-1]
    top_words = []
    for i in top_term_indices:
        top_words.append(features[i])
    topics[f"Topic {topic_idx + 1}"] = top_words

topics_df = pandas.DataFrame(topics)
print(f"{n_topics} topics x top {n_terms} terms each:")
print(topics_df)
```

```python
import scipy.cluster.hierarchy
import scipy.spatial.distance

dfm_dense = counts.astype(float)
row_totals = dfm_dense.sum(axis=1, keepdims=True)
row_totals[row_totals == 0] = 1.0
normalised = dfm_dense / row_totals

distances = scipy.spatial.distance.pdist(normalised, metric="euclidean")
linkage = scipy.cluster.hierarchy.linkage(distances, method="complete")

fig, ax = plt.subplots(figsize=(22, 7))
scipy.cluster.hierarchy.dendrogram(linkage, labels=labels, leaf_rotation=90, leaf_font_size=4, ax=ax)
ax.set_title("State of the Union (1790-2026) - Euclidean distance on normalised token frequency")
plt.tight_layout()
plt.show()
```

```python
import nltk.stem
import sklearn.feature_extraction.text

# 1980 onward. Stem the speeches before clustering, so word variants (govern,
# governs, government) collapse to one root.
stemmer = nltk.stem.SnowballStemmer("english")
modern_df = df[df["year"] >= 1980]
modern_texts = modern_df["text"].tolist()

stemmed_modern = []
for text in modern_texts:
    roots = []
    for word in text.lower().split():
        roots.append(stemmer.stem(word))
    stemmed_modern.append(" ".join(roots))

modern_vectorizer = sklearn.feature_extraction.text.CountVectorizer(stop_words="english", min_df=3)
modern_counts = modern_vectorizer.fit_transform(stemmed_modern).toarray()
modern_keep = modern_counts.sum(axis=0) >= 5
modern_counts = modern_counts[:, modern_keep]

modern_row_totals = modern_counts.sum(axis=1, keepdims=True).astype(float)
modern_row_totals[modern_row_totals == 0] = 1.0
modern_matrix = modern_counts / modern_row_totals

modern_labels = []
for year, pres in zip(modern_df["year"], modern_df["president"]):
    modern_labels.append(f"{year}-{pres}")

modern_distances = scipy.spatial.distance.pdist(modern_matrix, metric="euclidean")
modern_linkage = scipy.cluster.hierarchy.linkage(modern_distances, method="complete")

fig, ax = plt.subplots(figsize=(14, 6))
scipy.cluster.hierarchy.dendrogram(modern_linkage, labels=modern_labels, leaf_rotation=90, leaf_font_size=7, ax=ax)
ax.set_title("Modern SoUs (1980+) - Euclidean distance, stemmed")
plt.tight_layout()
plt.show()
```

<!-- demo slide -->

```python
cosine_distances = scipy.spatial.distance.pdist(modern_matrix, metric="cosine")
cosine_linkage = scipy.cluster.hierarchy.linkage(cosine_distances, method="complete")

fig, ax = plt.subplots(figsize=(14, 6))
scipy.cluster.hierarchy.dendrogram(cosine_linkage, labels=modern_labels, leaf_rotation=90, leaf_font_size=7, ax=ax)
ax.set_title("Modern SoUs (1980+) - Cosine distance, stemmed")
plt.tight_layout()
plt.show()
```

```python
founders_df = df[df["year"] < 1830]
founders_texts = founders_df["text"].tolist()

stemmed_founders = []
for text in founders_texts:
    roots = []
    for word in text.lower().split():
        roots.append(stemmer.stem(word))
    stemmed_founders.append(" ".join(roots))

founders_vectorizer = sklearn.feature_extraction.text.CountVectorizer(stop_words="english", min_df=3)
founders_counts = founders_vectorizer.fit_transform(stemmed_founders).toarray()
founders_keep = founders_counts.sum(axis=0) >= 5
founders_counts = founders_counts[:, founders_keep]

founders_row_totals = founders_counts.sum(axis=1, keepdims=True).astype(float)
founders_row_totals[founders_row_totals == 0] = 1.0
founders_matrix = founders_counts / founders_row_totals

founders_labels = []
for year, pres in zip(founders_df["year"], founders_df["president"]):
    founders_labels.append(f"{year}-{pres}")

founders_distances = scipy.spatial.distance.pdist(founders_matrix, metric="euclidean")
founders_linkage = scipy.cluster.hierarchy.linkage(founders_distances, method="complete")

fig, ax = plt.subplots(figsize=(18, 6))
scipy.cluster.hierarchy.dendrogram(founders_linkage, labels=founders_labels, leaf_rotation=90, leaf_font_size=6, ax=ax)
ax.set_title("Founders-era SoUs (pre-1830) - Euclidean distance, stemmed")
plt.tight_layout()
plt.show()
```
