## How to find data

- [Google Dataset Search](https://datasetsearch.research.google.com/)
- Twitter's Historical Powertrack: rolls the dice (and starts at US$5k)
  - [zenodo.org Twitter datasets](https://zenodo.org/search?q=twitter&file_type=zip&type=dataset)
  - Right now, only historical corpora matter
- [GLAM Workbench](https://glam-workbench.net/) — Trove API
- [data.gov.au](https://data.gov.au/home) and [data.nsw.gov.au](https://data.nsw.gov.au/)
- [YouTube Data API](https://developers.google.com/youtube/v3/docs/comments/list)
- Don't bother with Meta/Facebook
- Reddit is currently uncertain — [pushshift archive](https://paperswithcode.com/dataset/pushshift-reddit)
- [Papers With Code text datasets](https://paperswithcode.com/datasets?mod=texts)
- [Common Crawl](https://commoncrawl.org/)
- Dataverse — [Australian Data Archive](https://dataverse.ada.edu.au/)
- [Figshare](https://figshare.com/), [Dryad](https://datadryad.org/stash), [re3data.org](https://www.re3data.org/), [FAIRsharing](https://fairsharing.org/)
